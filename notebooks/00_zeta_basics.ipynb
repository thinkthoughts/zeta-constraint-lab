# 00 · Zeta Basics

A minimal starting notebook for **Zeta Constraint Lab**.

This notebook does four simple things:

1. defines the Riemann zeta function numerically  
2. samples $\zeta(s)$ on a small region of the complex plane  
3. visualizes $\log(1 + |\zeta(s)|)$  
4. samples the critical line $\Re(s)=\frac12$ and marks the first few known nontrivial zeros

## VC / IA framing

This notebook uses a **VC-style** setup:

- standard object: $\zeta(s)$  
- explicit domain in the complex plane  
- reproducible numerical evaluation  
- critical-line sampling treated as a defined constraint boundary  

It does **not** claim a theorem or proof of RH.

```python
import numpy as np
import mpmath as mp
import matplotlib.pyplot as plt

mp.mp.dps = 30
```

## Define $\zeta(s)$ and helper functions

```python
def zeta(s):
    return mp.zeta(s)

def zeta_abs(x, y):
    s = x + 1j * y
    return abs(zeta(s))

def safe_log1p_abs_zeta(x, y):
    return float(mp.log(1 + zeta_abs(x, y)))
```

## Sample a small region of the complex plane

We evaluate on a modest grid so the notebook stays lightweight and Colab-friendly.

```python
x_min, x_max = -2.0, 2.0
y_min, y_max = 0.0, 35.0

nx, ny = 180, 220
xs = np.linspace(x_min, x_max, nx)
ys = np.linspace(y_min, y_max, ny)

Z = np.empty((ny, nx), dtype=float)

for i, y in enumerate(ys):
    for j, x in enumerate(xs):
        Z[i, j] = safe_log1p_abs_zeta(x, y)

Z.min(), Z.max()
```

## Heatmap of $\log(1 + |\zeta(s)|)$

Dark valleys tend to indicate regions near zeros.
The dashed vertical line marks the critical line $\Re(s)=\frac12$.

```python
plt.figure(figsize=(8, 6))
plt.imshow(
    Z,
    extent=[x_min, x_max, y_min, y_max],
    origin="lower",
    aspect="auto"
)
plt.colorbar(label=r"$\log(1 + |\zeta(s)|)$")
plt.axvline(0.5, linestyle="--", linewidth=1, label=r"$\Re(s)=1/2$")
plt.xlabel(r"$\Re(s)$")
plt.ylabel(r"$\Im(s)$")
plt.title(r"Riemann zeta function: $\log(1 + |\zeta(s)|)$")
plt.legend()
plt.show()
```

## Sample the critical line

Now we restrict to the line

\[
s = \frac12 + it.
\]

This is the main constraint boundary used in the repo.

```python
t_vals = np.linspace(0, 35, 1500)
zeta_on_line = np.array([abs(zeta(0.5 + 1j * t)) for t in t_vals], dtype=float)

known_zeros = np.array([
    14.134725141,
    21.022039639,
    25.010857580,
    30.424876126,
    32.935061588
])
```

```python
plt.figure(figsize=(8, 4.8))
plt.plot(t_vals, zeta_on_line, linewidth=1.5, label=r"$|\zeta(1/2 + it)|$")
for idx, tz in enumerate(known_zeros):
    plt.axvline(tz, linestyle="--", linewidth=1)
plt.yscale("log")
plt.xlabel(r"$t$")
plt.ylabel(r"$|\zeta(1/2 + it)|$")
plt.title(r"Critical-line sampling: $|\zeta(1/2 + it)|$")
plt.legend()
plt.show()
```

## Local view near the first nontrivial zero

This makes the first valley easier to inspect numerically.

```python
t0 = known_zeros[0]
local_t = np.linspace(t0 - 1.0, t0 + 1.0, 600)
local_vals = np.array([abs(zeta(0.5 + 1j * t)) for t in local_t], dtype=float)

plt.figure(figsize=(8, 4.5))
plt.plot(local_t, local_vals, linewidth=1.5)
plt.axvline(t0, linestyle="--", linewidth=1, label=f"first zero ≈ {t0:.6f}")
plt.yscale("log")
plt.xlabel(r"$t$")
plt.ylabel(r"$|\zeta(1/2 + it)|$")
plt.title(r"Local view near the first nontrivial zero")
plt.legend()
plt.show()
```
